Определить типы переменных

* номинальные переменные (не имеют порядка и сравнения)
    1. school_name - названия школ
    2. state - названия штатов

* порядковые переменные (имеют порядок, но не имеют смысла мат операции)

    1. school_type - Типы школ (Public, Private, Charter)
    2. grade_level - Уровень школы (Elementary, Middle, High)

* числовые непрерывные переменные (имеют порядок, есть смысл мат операций)
    1. funding_per_student_usd - Ежегодное **финансирование** на одного ученика в долларах
    2. avg_test_score_percent - Средний **балл** успеваемости учащихся (0–100%).
    3. student_teacher_ratio - Среднее **количество** учеников на одного учителя
    4. percent_low_income - **Процент** учащихся из малообеспеченных семей
    5. percent_minority - **Процент** учащихся из меньшинств
    6. internet_access_percent - **Процент** учащихся, имеющих доступ к интернету
    7. dropout_rate_percent - Ежегодный **процент** отсева учащихся

````python
df['school_type'] = df['school_type'].astype('category')
df['grade_level'] = df['grade_level'].astype('category')
````

## **Визуализация распределения переменных**

### **Описание:**
Визуализации для понимания распределения числовых и категориальных переменных.



1. **Гистограмма для числовых переменных:**

````python
num_cols = df.select_dtypes(include='float').columns
fig, axes = plt.subplots(1, len(num_cols), figsize=(6 * len(num_cols), 4))

for ax, col in zip(axes, num_cols):
   sns.histplot(df[col], kde=True, ax=ax)
   ax.set_title(f'Распределение {col}')
   ax.set_xlabel(col)
   ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
````

````python
hue_cols = df.select_dtypes(include='category').columns
fig, axes = plt.subplots(len(hue_cols), len(num_cols), figsize=(6 * len(num_cols), 8))

for i, hue_col in enumerate(hue_cols):
    for j, num_col in enumerate(num_cols):
        ax = axes[i, j]
        sns.histplot(data=df, x=num_col, kde=True, ax=ax,hue=hue_col)
        ax.set_title(f'{num_col} по {hue_col}')
        ax.set_xlabel(num_col)
        ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
````

Описание графиков:
* **funding_per_student_usd**
    - Не похож на нормальное распределение (напоминает обратную картину нормального распределения)
    - большинство значений сконцентрированно или вокруг одного из двух значений
* **avg_test_score_percent**
    - Также не похоже на нормальное распределение, если смотреть на общий график или по степени образования, но если смотреть на типы школ, у Частных и Публичных школ нормальное распределение успеваемости, для Специальных школ опять характерна "вилка" для среднего значения
    - Большинство значений сконцентрированно или вокруг 85-90% средней успеваемости учеников или 45%
* **student_teacher_ratio**
    - Не нормальное распределение
    - Видны сосредоточения на 15 или 26 учениках на учителя
* **percent_low_income**
    - Похоже на нормальное распределение
    - Концентрация находится где-то на 65%, но при рассмотрении по категориям можно увидеть, что для каждой категории характерно свое значение
* **percent_minority**
    - Нормальное распределение
    - Большинство значений концентрируется вокруг 50%
* **internet_access_percent**
    - На некоторых категориях есть какие-то подобия нормального распределения
    - В зависимости от категории значения концентрируются вокруг разных значений
* **dropout_rate_percent**
    - Не нормальное распределение
    - В основном значения концентрируются или вокруг 4.5% или вокруг 12.5%

**Постройте корреляционную матрицу для всех числовых переменных и визуализируйте её с помощью тепловой карты:**

````python
# Корреляционная матрица
correlation_matrix = df.select_dtypes('float64').corr()

# Визуализация тепловой карты
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin = -0.05, vmax = 0.05)
plt.title('Корреляционная матрица')
plt.show()
````

````python
category_column = 'school_type'

categories = df[category_column].unique()

fig, axes = plt.subplots(1, len(categories), figsize=(6 * len(categories), 5))

for ax, category in zip(axes, categories):
    category_data = df[df[category_column] == category]
    correlation_matrix = category_data.select_dtypes('float64').corr()
    sns.heatmap(correlation_matrix,
                annot=True,
                cmap='coolwarm',
                vmin=-0.1,
                vmax=0.1,
                ax=ax,)

    ax.set_title(f'Корреляция: {category}')

plt.tight_layout()
plt.show()
````
